```python
def max3(a, b, c):
  """ max3 test sample"""
  maximum = a
  if b > maximum: maximum = b
  if c > maximum: maximum = c
  return maximum

print(f'max3(3, 2, 1)= {max3(3, 2, 1)}')
print(f'max3(3, 2, 2)= {max3(3, 2, 2)}')
print(f'max3(3, 1, 2)= {max3(3, 1, 2)}')
print(f'max3(3, 2, 3)= {max3(3, 2, 3)}')
print(f'max3(1, 2, 3)= {max3(1, 2, 3)}')
print(f'max3(2, 3, 1)= {max3(2, 3, 1)}')
```

```python
def med3(a, b, c):
  """a, b, c median value"""

  if a >= b:
    if b >= c:
      return b
    elif a <= c:
      return a
    else:
      return c
  elif a > c:
    return a
  elif b > c:
    return c
  else:
    return b

print("median value")    
a = int(input('int a input=>'))
b = int(input('int b input=>'))
c = int(input('int c input=>'))
print(f'median value={med3(a, b, c)}')

```

Operator *를 통한 Packing 

operator *는 the tuple (or iterable) unpacking operator

```python
*a, = 1, 2
print(a)
print('\n')

a, *b = 1, 2, 3
print(a)
print(b)
print('\n')

*a, b = 1, 2, 3
print(a)
print(b)
print('\n')

a, *b, c = 1, 2, 3
print(a)
print(b)
print(c)
print('\n')

a, *b, c, d = 1, 2, 3
print(a)
print(b)
print(c)
print(d)
print('\n')
```

```python
gen = (2 ** x for x in range(10))
print(gen)

*g, = gen
print(g)

```

```python
*gen,  = (2 ** x for x in range(10))  # tuple로 값을 받을수 있다
print(gen)
```

함수의 튜블리턴(Returning Tuples in Functions)
[출처] 파이썬 iterable unpacking,packing|작성자 용용

```python
def powers(number):
  return number, number ** 2, number ** 3
  # packing returned values in a tuple
result = powers(2)  
print(result)
  
  # unpacking returned values to multiple variables
number, square, cube = powers(2)
print(number, square, cube)

*_, cube = powers(2)
print(cube)
```
