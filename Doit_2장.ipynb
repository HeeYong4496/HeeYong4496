Any는 제약없는 임의의 자료형, Sequence는 시퀀스형을 의미, 리스트형, 바이트 배열형, 문자열형, 튜플형, 바이트형

```python
from typing import Any, Sequence

def max_of(a: Sequence) -> Any:
     # 건네 받은 매개변수 a의 자료형은 Sequence이다
     # 시퀀스형 a 원소의 최대값을 반환하는 것은 Any
    maxn = a[0]
    for i in range(1, len(a)):
        if a[i] > maxn:
            maxn = a[i]
    return maxn

if __name__ == '__main__':
    print('배열의 최대값을 구함')    
    num = int(input('원소 수를 입력='))

    x = [None] * num

    for i in range(num):
        x[i] = int(input(f'x[{i}] 값을 입력'))

    print(f'최대값 {max_of(x)} .')
```

```python
from typing import Any, MutableSequence

def reverse_array(a: MutableSequence) -> None:

    n = len(a)
    for i in range(n // 2):
        a[i], a[n - i -1] = a[n - i -1], a[i]
if __name__ == '__main__':
    print('역순 정렬') 
    nx = int(input('원소 수를 입력'))
    x = [None] * nx

    for i in range(nx):
        x[i] = int(input(f'x[{i}] 입력='))
    reverse_array(x)

    print('정렬')
    for i in range(nx):
        print(f'x[{i}] = {x[i]}')
```
