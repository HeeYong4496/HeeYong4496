```python
from max import max_of

print('배열의 최대값 구함')
print('주의: "End" 입력하면 종료')

number = 0
x = []

while True:
    s = input(f'x[{number}] 값을 입력:')
    if s == 'End':
        break
    x.append(int(s))
    number += 1

print(f'{number}개를 입력')    
print(f'최대값은 {max_of(x)} 이다')    
```
