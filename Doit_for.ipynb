```python
area = int(input('직사각형의 넓이 입력=>'))

for i in range(1, area + 1):
    if i * i > area: break
    if area % i: continue
    print(f'{i} x {area // i}')
```

```python
import random

n = int(input('난수의 갯수 입력='))

for _ in range(n):
    r = random.randint(10, 99)
    print(r, end=' ')
    if r == 13:
        print('\n stop')
        break
else:
    print('\n 난수 종료')  
    
          
```

```python
n = int(input('짧은 변의 길이 입력='))

for i in range(n):
    for j in range(i + 1):
        print('*', end='')
    print()
```

```python
n = int(input('짧은 변의 길이 입력='))

for i in range(n):
    for _ in range(n - i - 1):
        print(' ', end='')
    for _ in range(i + 1):
        print('*', end='')
    print()
```
